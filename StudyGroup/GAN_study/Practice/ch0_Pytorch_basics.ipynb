```python
import torch
```

```python
# normal python variables
x = 3.5
y = x*x + 2

print(x,y)
```

```python
# simple pytorch tensor
x = torch.tensor(3.5)
print(x)
print(type(x))
```

```python
# simple arithmetic with tensors

y = x+3
print(y)
print(type(y))
```

```python
# pytorch tensor with gradient enabled
x = torch.tensor(3.5, requires_grad=True)
print(x)

# x로부터 정의된 y
y = (x-1) * (x-2) * (x-3)
print(y)
```

```python
# 기울기 계산
y.backward()
```

```python
# what is gradient at x=3.5
x.grad
```

```python
# x,y,z에 관련된 간단한 그래프 설정
x = torch.tensor(3.5, requires_grad = True)
y = x*x
z = 2*y + 3

# 기울기 계산
z.backward()
print(x.grad)
```

```python
# set up simple graph relating x,y and z
a = torch.tensor(2.0, requires_grad = True)
b = torch.tensor(1.0, requires_grad = True)

x = 2*a + 3*b
y = 5*a*a + 3*b*b
z = 2*x + 3*y

# work out gradients
z.backward()

# what is gradient at a = 2.0
a.grad
```
